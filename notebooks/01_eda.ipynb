# 01 - Exploratory Data Analysis (EDA)

In this notebook, we'll perform EDA on the Heart Disease dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Adjust plot settings
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

### 1. Load Data


```python
# Load the CSV
data_path = '../data/heart.csv'
if os.path.exists(data_path):
    df = pd.read_csv(data_path)
    print(f"Shape: {df.shape}")
    print("\nDtypes:\n", df.dtypes)
    display(df.head())
else:
    print(f"Dataset not found at {data_path}. Please download it from Kaggle and place it in the data/ folder.")
```

### 2. Target Column Analysis


```python
if 'df' in locals():
    print(df['HeartDisease'].value_counts())
    sns.countplot(x='HeartDisease', data=df)
    plt.title('Target Variable Distribution (HeartDisease)')
    plt.show()
```

### 3. Identify Column Types


```python
if 'df' in locals():
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
    numeric_cols = df.select_dtypes(exclude=['object']).columns.tolist()
    numeric_cols.remove('HeartDisease') # Exclude target from features
    print("Categorical Columns:", categorical_cols)
    print("Numeric Columns:", numeric_cols)
```

### 4. Numeric Features Analysis


```python
if 'df' in locals():
    display(df[numeric_cols].describe())
    
    # Plot histograms
    for col in numeric_cols:
        plt.figure(figsize=(6,4))
        sns.histplot(df[col], kde=True)
        plt.title(f'Histogram of {col}')
        plt.show()
```

### 5. Categorical Features Analysis


```python
if 'df' in locals():
    for col in categorical_cols:
        print(f"\n--- {col} ---")
        print(df[col].value_counts())
        plt.figure(figsize=(6,4))
        sns.countplot(x=col, data=df, hue='HeartDisease')
        plt.title(f'Countplot of {col}')
        plt.show()
```

### 6. Correlation Heatmap


```python
if 'df' in locals():
    plt.figure(figsize=(10, 8))
    corr_matrix = df[numeric_cols + ['HeartDisease']].corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap')
    plt.show()
```

### 7. Boxplots grouped by HeartDisease


```python
if 'df' in locals():
    for col in numeric_cols:
        plt.figure(figsize=(6,4))
        sns.boxplot(x='HeartDisease', y=col, data=df)
        plt.title(f'Boxplot of {col} by HeartDisease')
        plt.show()
```

### 8. Summary of Findings
- Are there class imbalances? No, the target is relatively balanced.
- Which features look most predictive? Features like `Oldpeak`, `MaxHR`, and `Age` show noticeable differences between patients with and without heart disease in the boxplots.
- Are there outliers? Yes, e.g., in `Cholesterol`, there are 0 values which are physiologically impossible. `RestingBP` also has some 0 values.
